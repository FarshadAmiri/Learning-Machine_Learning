```python
face = photo[70:395,260:505]
row_counter = -1
column_counter = -1
for row in face:
  row_counter += 1
  for cell in row:
    if column_counter == 244:
      column_counter = -1
    column_counter += 1
    if ((cell[0]>100) or (cell[1]>100) or (cell[2]>100)):
      face[row_counter][column_counter] = [255,255,255]
plt.imshow(face)
```
